This project starts with data set up where the data of Austin, Texas are getting from API. After the preliminary analysis of the data, data cleaning and comparing are done.

## Importing Data

```python
import pandas as pd
from sodapy import Socrata

#Getting data from API
client = Socrata("data.austintexas.gov", "vT9dyEXWAk1UrzgBe8KXFuaU9")

# First 3000000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("7d8e-dm7r", limit=3000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

## Cleaning the Data

```python
# cleaning data
results_df.dropna(axis='rows')

def get_empty_rows(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'nan':
                if i not in indices:
                    indices.append(i)

    return indices

def get_empty_rows2(results_df):
    indices = []
    for i, s in results_df.iterrows():
        for v in s.values:
            if str(v).lower() == 'null':
                if i not in indices:
                    indices.append(i)

    return indices

list_of_empty_rows = get_empty_rows(results_df)
list_of_empty_rows2 = get_empty_rows(results_df)

deleted = list_of_empty_rows + list_of_empty_rows2

df = results_df.drop(deleted)
```

### Number of unique vehicles by year and vehicle type

```python
#number of unique vehicles by month and year and vehicles type(with data)
records_year_type = df.groupby(["date",  "vehicle_type"])['device_id'].nunique().reset_index(name='count')
records_year_type
```
